```python
import threading
import os
import time
```

```python
def one(x):
    print("Waiting 10 minutes to terminate for", x)
    
    for i in x:
#         time.sleep(0.5) # at each iteration it waits for 0.3 time
        print("The number is: ", i)


def two(y):
    print("Waiting 5 minutes to terminate next", y)
    
    for j in y:
#         time.sleep(0.3) # at each iteration it waits for 0.3 time
        print("The next number is: ", j)
        
        
a1 = [5, 10, 15]
a2 = [2, 4, 5]
```

```python
t = time.time() # get total time to execute the functions 

t1 = threading.Thread(target=one, args= (a1,))
t2 = threading.Thread(target=two, args= (a2,))
t2.start()
# t2.join()
t2.join(5)        #What 5 represents here?? Find that out.
t1.start()
t1.join()
# t1.join(10)

print("\nTotal time taking by threads is :", time.time() - t) # print the total time  
```
