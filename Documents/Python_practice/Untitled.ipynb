```python
import time
```

```python
start_time = time.time()

count, total_count = 0, 0

for i in range(0, 2):
    p, Cr_sub = 0, 0
    
    for j in range(2, 4):
        print("\nFor the {}th student".format(i+1))
        print("{}th subject credit is: ".format(j) , j)
        if (Cr_sub +j) <= i:    
            count+= 1
            print("count: ", count)
            
        print("subject counts: ", count)
            
        Cr_sub+= p
        p+= 1
    
        print("Subjects credit added: ", Cr_sub)
    total_count += count
    print("Total count: ", total_count, "\n")
        
print("\nNumber of ordered pairs are: ", count, total_count)
print("Total number of subject are: ", p, " and the total credits of the subjects is: ", Cr_sub)

print("--- %s seconds ---" % (time.time() - start_time))
```
