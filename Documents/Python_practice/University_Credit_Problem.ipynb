```python
def creating_tuple(n):
    t = []
    for i in range(0, n):
        x = int(input("Enter it's value: "))
        t.append(x)
    tpl = tuple(t)
    return tpl
```

```python
k = int(input("Enter the no. of Subjects: "))
print("Enter the credits of the {} subjects: " .format(k))
Subs_cr = creating_tuple(k)

print(Subs_cr)
# print(Subs_cr[0])
# print(k)
```

```python
from itertools import chain, combinations

def powerset(iterable):
    "powerset([1,2,3]) --> () (1,) (2,) (3,) (1,2) (1,3) (2,3) (1,2,3)"
    s = list(iterable)
    return chain.from_iterable(combinations(s, r) for r in range(len(s)+1))

x = list(powerset(Sub_cr))
x.remove(())
print("\nAll the combinations of the subjects credits are: \n",
      x, "\nand the number of elements in it:", len(x))
```
